# 12. Classify the employees into 3 categories.
# Category 1 : employees below 18.5 bmi - underweight
# Category 2 : employees with bmi in the range of 18.5 to 24.9 - fit
# Category 3 : employee with bmi over 24.9 - overweight
# What is the average expenditure of employees in each category.
# Display the bar chart for the same


```python
list1 = []
for i in df_emp_health.bmi:
    if i < 18.5:
        list1.append('underweight')
    elif i >= 18.5 and i < 24.9 :
        list1.append('fit')
    else :
        list1.append('overweight')
        
df_emp_health.insert(loc=len(df_emp_health.columns),column = 'bmi_category',value=list1)        

df_emp_health
```

```python
df_emp_health.groupby(['bmi_category']).expenditure.mean()
```

```python
plt.figure(figsize=(15,8))
plt.title('Expenditure for  employees based on all 3 category of BMI')
sns.barplot(x = df_emp_health.bmi_category, y = df_emp_health.expenditure)
plt.show()
# Below is bar chart for BMI for  employees based on all 3 category 
# Expenditure is higher for employees who is overweight and lowest for employees who is underweight
```
